```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tf_ratio = pd.read_csv('results/bert_ext/transformer_ratio.csv')
blob_ratio = pd.read_csv('results/bert_ext/blob_ratio.csv')
unfiltered_ratio = pd.read_csv('results/bert_ext/unfiltered_ratio.csv')

tf_ratio = tf_ratio.sort_values('parameters/summary_ratio')
blob_ratio = blob_ratio.sort_values('parameters/summary_ratio')
unfiltered_ratio = unfiltered_ratio.sort_values('parameters/summary_ratio')
```

```python
fig, ax = plt.subplots(1, 3)
fig.set_figwidth(13)



ax[0].plot(tf_ratio['parameters/summary_ratio'], tf_ratio['rouge1'], label='transformer')
ax[0].plot(blob_ratio['parameters/summary_ratio'], blob_ratio['rouge1'], label='blob') 
ax[0].plot(unfiltered_ratio['parameters/summary_ratio'], unfiltered_ratio['rouge1'], label='unfiltered')

ax[1].plot(tf_ratio['parameters/summary_ratio'], tf_ratio['rouge2'], label='transformer')
ax[1].plot(blob_ratio['parameters/summary_ratio'], blob_ratio['rouge2'], label='blob') 
ax[1].plot(unfiltered_ratio['parameters/summary_ratio'], unfiltered_ratio['rouge2'], label='unfiltered')

ax[2].plot(tf_ratio['parameters/summary_ratio'], tf_ratio['results/rougeL'], label='transformer')
ax[2].plot(blob_ratio['parameters/summary_ratio'], blob_ratio['results/rougeL'], label='blob') 
ax[2].plot(unfiltered_ratio['parameters/summary_ratio'], unfiltered_ratio['results/rougeL'], label='unfiltered')
handles, labels = ax[2].get_legend_handles_labels()
ax[0].set_title('Rouge1')
ax[1].set_title('Rouge2')
ax[2].set_title('RougeL')
ax[1].set_xlabel('Summary ratio')
ax[0].set_ylabel('Score')
fig.legend(handles, labels, title='Filtering method')
fig.suptitle('BertExt results for different summary ratio on newsroom')
fig.savefig('plots/bertExt_summary_ratio_newsroom.png')
```

```python
multinews_ratio = pd.read_csv('results/bert_ext/multinews_ratio.csv')
```

```python
multinews_tf_ratio = multinews_ratio[multinews_ratio['parameters/threshold'] == 0.001].sort_values('parameters/summary_ratio')
multinews_blob_ratio = multinews_ratio[multinews_ratio['parameters/threshold'] == 0.4].sort_values('parameters/summary_ratio')
multinews_unfiltered_ratio = multinews_ratio[multinews_ratio['parameters/threshold'] == 0.01].sort_values('parameters/summary_ratio')
```

```python
fig, ax = plt.subplots(1, 3)
fig.set_figwidth(13)



ax[0].plot(multinews_tf_ratio['parameters/summary_ratio'], multinews_tf_ratio['rouge1'], label='transformer')
ax[0].plot(multinews_blob_ratio['parameters/summary_ratio'], multinews_blob_ratio['rouge1'], label='blob') 
ax[0].plot(multinews_unfiltered_ratio['parameters/summary_ratio'], multinews_unfiltered_ratio['rouge1'], label='unfiltered')

ax[1].plot(multinews_tf_ratio['parameters/summary_ratio'], multinews_tf_ratio['rouge2'], label='transformer')
ax[1].plot(multinews_blob_ratio['parameters/summary_ratio'], multinews_blob_ratio['rouge2'], label='blob') 
ax[1].plot(multinews_unfiltered_ratio['parameters/summary_ratio'], multinews_unfiltered_ratio['rouge2'], label='unfiltered')

ax[2].plot(multinews_tf_ratio['parameters/summary_ratio'], multinews_tf_ratio['results/rougeL'], label='transformer')
ax[2].plot(multinews_blob_ratio['parameters/summary_ratio'], multinews_blob_ratio['results/rougeL'], label='blob') 
ax[2].plot(multinews_unfiltered_ratio['parameters/summary_ratio'], multinews_unfiltered_ratio['results/rougeL'], label='unfiltered')
handles, labels = ax[2].get_legend_handles_labels()
ax[0].set_title('Rouge1')
ax[1].set_title('Rouge2')
ax[2].set_title('RougeL')
ax[1].set_xlabel('Summary ratio')
ax[0].set_ylabel('Score')
fig.legend(handles, labels, title='Filtering method')
fig.suptitle('BertExt results for different summary ratio on multinews')
fig.savefig('plots/bertExt_summary_ratio_multinews.png')
```
